## **Content**
This data set consists of three types of entities: (a) the specification of an auto in terms of various characteristics, (b) its assigned insurance risk rating, (c) its normalized losses in use as compared to other cars. The second rating corresponds to the degree to which the auto is more risky than its price indicates. Cars are initially assigned a risk factor symbol associated with its price. Then, if it is more risky (or less), this symbol is adjusted by moving it up (or down) the scale. Actuarians call this process "symboling". A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe.

The third factor is the relative average loss payment per insured vehicle year. This value is normalized for all autos within a particular size classification (two-door small, station wagons, sports/speciality, etcâ€¦), and represents the average loss per car per year.

Note: Several of the attributes in the database could be used as a "class" attribute.

```python
# remove all dash '-' from columns with replace to underscore '_' and convert all names of columns to lower case

Automobile_data.columns = Automobile_data.columns.str.replace('-','_').str.lower()
Automobile_data.columns
```

```python
# We can notice here that we have a fixed range of numbers starting from 3 to -2, so we convert data type from int to category

print(Automobile_data['symboling'].value_counts())
print('\nThe sum of values: ' + str(Automobile_data['symboling'].value_counts().sum()))
```

```python
#function to get some information for all columns in dataframe as unique value, number of value and data type

def info_columns(dataframe): 
   
    for column in dataframe.columns:
        print('Name of columns: '+ column + '\n')
        print(dataframe[column].unique())
        print('\nThe sum of values: ' + str(dataframe[column].value_counts().sum()))
        print('and data type is: '+ str(dataframe[column].dtype))
        print('\n--------------------------------------\n')
        
        
# 1- We can notice here that we have some columns have a fixed range of numbers, so we convert data type to category
# 2- And we notice some data type invalid because the columns have a '?' that This is the main reason for converting 
# columns to object data type
# 3- we have some missing value but not explicitly the missing value represent as '?'


info_columns(Automobile_data)
```

```python
#Here two columns from columns that will convert data type to category the 'num_of_doors' have a missing value id '?' 
#repeated two times

print(Automobile_data['num_of_doors'].value_counts())
print('\n--------------------------------------\n')
Automobile_data['num_of_doors'] = Automobile_data['num_of_doors'].replace('?', 'not_found')

Automobile_data['num_of_doors'].unique()
```

```python
#change data type of columns that which should be "category" data type

category_columns = ['symboling', 'fuel_type', 'aspiration', 'num_of_doors', 'body_style', 'drive_wheels', 'engine_location',
                    'engine_type', 'num_of_cylinders', 'fuel_system']

for column in category_columns:
    Automobile_data[column] = Automobile_data[column].astype('category')
    
Automobile_data.info()
```

### Transform (normalized-losses, bore, stroke, horsepower, peak_rpm, price) from object to numeric
